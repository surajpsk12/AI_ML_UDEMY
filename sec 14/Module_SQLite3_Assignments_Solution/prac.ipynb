```python
import sqlite3
```

```python
connection = sqlite3.connect("example.db")
connection
```

```python
cursor = connection.cursor()
```

```python
##create a table 
cursor.execute('''
    create table if not exists employees(
        id Integer Primary Key,
        name Text not null,
        age Integer , 
        dept Text 
    )
''')
connection.commit()
```

```python
## insertion 
cursor.execute('''
    insert into employees(name,age,dept)
        values("Suraj",22,'sde')
    
''')
cursor.execute('''
    insert into employees(name,age,dept)
        values("dhiraj",22,'sdde')
    
''')
cursor.execute('''
    insert into employees(name,age,dept)
        values("Sufa",202,'sded')
    
''')

connection.commit()
```

```python
## read the data 
cursor.execute("select * from employees")
rows=cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
    delete from employees where name="Suraj"
''')

connection.commit()
```
