```python
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def __init__(self,r):
        self.r=r

    def area(self):
        print(f"the area of circle is {3.14*self.r*self.r}")

class Square(Shape):
    def __init__(self,side):
        self.side=side

    def area(self):
        print(f"the area of square is {self.side*self.side}")    
    
shape1 = Circle(1).area()
shape2 = Square(2).area()
```

```python
from abc import ABC,abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    
    def start_engine(self):
        print(f"engine started by car")
    
class Bike(Vehicle):

    def start_engine(self):
        print(f"engine started by bike")

Car().start_engine()
Bike().start_engine()
```
